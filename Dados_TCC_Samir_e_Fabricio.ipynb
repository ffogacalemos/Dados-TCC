```python
#ATIVIDADE 1 e 2 aceitação de tecnologia ChatGPT
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.lines import Line2D

# Dados CSV fornecidos adaptados ao exemplo
data = {
      "Pergunta": [
        "Eu acredito que o ChatGPT forneceu\n resultados úteis para mim.",
        "Foi fácil interagir com o ChatGPT.",
        "Eu sinto que tenho\n controle sobre a interação com o chatGPT.\n Sinto que consegui formular\n perguntas de maneira clara e concisa.",
        "Estou satisfeito(a) com o desempenho\n do ChatGPT, pois ele forneceu\n bons resultados e me ajudou a resolver\n tarefas de forma eficiente.",
        "Me senti confiante em relação a minha \nhabilidade em interagir com o ChatGPT.\n Sinto que ele me possibilitou\n melhor domínio e compreensão da atividade.",
        "Durante o experimento, tive os recursos\n necessários para completá-lo de maneira\n eficiente, incluindo\n uma conexão de internet estável,\n computadores de qualidade e acesso\n ao ChatGPT adequadamente."
    ],
    "Discordo Totalmente":       [0, 0, 0, 0, 0, 0],
    "Discordo":                  [0, 1, 1, 1, 1, 1],
    "Não concordo nem discordo": [2, 1, 2, 3, 1, 3],
    "Concordo":                  [8, 11, 11, 5, 11, 5],
    "Concordo Totalmente":       [8, 4, 3, 8, 4, 7]

}

# Convertendo em DataFrame
df = pd.DataFrame(data)

# Calculando as porcentagens
df_perc = df.set_index('Pergunta').apply(lambda x: x / x.sum() * 100, axis=1)

# Configurando o gráfico
fig, ax = plt.subplots(figsize=(12, 8))

# Criando o gráfico de barras divergentes
for i, (pergunta, row) in enumerate(df_perc.iterrows()):
    left = -row['Discordo Totalmente'] - row['Discordo'] - row['Não concordo nem discordo']/2
    ax.barh(pergunta, row['Discordo Totalmente'], color='darkred', left=left)
    ax.barh(pergunta, row['Discordo'], color='lightcoral', left=left + row['Discordo Totalmente'])
    ax.barh(pergunta, row['Não concordo nem discordo'], color='grey', left=left + row['Discordo Totalmente'] + row['Discordo'])
    ax.barh(pergunta, row['Concordo'], color='lightgreen', left=left + row['Discordo Totalmente'] + row['Discordo'] + row['Não concordo nem discordo'])
    ax.barh(pergunta, row['Concordo Totalmente'], color='darkgreen', left=left + row['Discordo Totalmente'] + row['Discordo'] + row['Não concordo nem discordo'] + row['Concordo'])

    # Adicionando as porcentagens dentro das barras, ajustando as posições para evitar sobreposição
    if row['Discordo Totalmente'] > 0:
        ax.text(left + row['Discordo Totalmente']/2, i, f"{row['Discordo Totalmente']:.1f}%", ha='center', va='center', color='white')
    if row['Discordo'] > 0:
        ax.text(left + row['Discordo Totalmente'] + row['Discordo']/2, i, f"{row['Discordo']:.1f}%", ha='center', va='center', color='black')
    if row['Não concordo nem discordo'] > 0:
        ax.text(left + row['Discordo Totalmente'] + row['Discordo'] + row['Não concordo nem discordo']/2, i, f"{row['Não concordo nem discordo']:.1f}%", ha='center', va='center', color='black')
    if row['Concordo'] > 0:
        ax.text(left + row['Discordo Totalmente'] + row['Discordo'] + row['Não concordo nem discordo'] + row['Concordo']/2, i, f"{row['Concordo']:.1f}%", ha='center', va='center', color='black')
    if row['Concordo Totalmente'] > 0:
        ax.text(left + row['Discordo Totalmente'] + row['Discordo'] + row['Não concordo nem discordo'] + row['Concordo'] + row['Concordo Totalmente']/2, i, f"{row['Concordo Totalmente']:.1f}%", ha='center', va='center', color='white')

# Adicionando uma linha vertical no centro do gráfico
ax.axvline(0, color='black', linewidth=0.8)


# Adicionando a legenda
legend_labels = ['Discordo Totalmente', 'Discordo', 'Não concordo nem discordo', 'Concordo', 'Concordo Totalmente']
legend_colors = ['darkred', 'lightcoral', 'grey', 'lightgreen', 'darkgreen']

# Criando a legenda personalizada
legend_lines = [Line2D([0], [0], color=color, lw=4) for color in legend_colors]
ax.legend(
    handles=legend_lines,
    labels=legend_labels,
    loc='lower center',
    bbox_to_anchor=(0.5, -0.1),  # Ajustado para estar mais perto do gráfico
    ncol=5,
    frameon=False,
    fontsize=10,
    title_fontsize='12'
)

# Mostrando o gráfico
plt.show()
```

```python

# Substituir vírgulas por pontos e converter as colunas para tipo numérico
data = data.applymap(lambda x: str(x).replace(',', '.') if isinstance(x, str) else x)
data = data.apply(pd.to_numeric, errors='coerce')

# Renomear colunas para facilitar a manipulação
data.columns = [
    "Controle_Exp1", "GPT_Exp1", "Controle_Exp2", "GPT_Exp2",
    "Pré_Controle", "Pré_GPT", "Controle_Exp3", "GPT_Exp3",
    "Controle_Ativ2_Exp3", "GPT_Ativ2_Exp3", "Pós_Controle", "Pós_GPT"
]

# Criar gráficos boxplot para as solicitações
import matplotlib.pyplot as plt
import seaborn as sns

# Configuração de estilo
sns.set(style="whitegrid")

# Gráfico 1: Atividades 1, 2 e 3 para os grupos Controle e GPT
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Preparar os dados para o primeiro gráfico
atividade_data = pd.DataFrame({
    "Grupo": ["Controle"] * 3 * len(data) + ["Experimental"] * 3 * len(data),
    "Atividade": ["Atividade 1"] * len(data) + ["Atividade 2"] * len(data) + ["Atividade 3"] * len(data) +
                 ["Atividade 1"] * len(data) + ["Atividade 2"] * len(data) + ["Atividade 3"] * len(data),
    "Notas": pd.concat([data["Controle_Exp1"], data["Controle_Exp2"], data["Controle_Exp3"],
                        data["GPT_Exp1"], data["GPT_Exp2"], data["GPT_Exp3"]], ignore_index=True)
})

sns.boxplot(ax=axes[0], data=atividade_data, x="Atividade", y="Notas", hue="Grupo")
axes[0].set_title("Atividades 1, 2 e 3 por Grupo")
axes[0].set_ylabel("Notas")
axes[0].set_xlabel("Atividades")

# Gráfico 2: Pré-teste e Pós-teste para os grupos Controle e GPT
teste_data = pd.DataFrame({
    "Grupo": ["Controle"] * 2 * len(data) + ["Experimental"] * 2 * len(data),
    "Teste": ["Pré-teste"] * len(data) + ["Pós-teste"] * len(data) +
             ["Pré-teste"] * len(data) + ["Pós-teste"] * len(data),
    "Notas": pd.concat([data["Pré_Controle"], data["Pós_Controle"],
                        data["Pré_GPT"], data["Pós_GPT"]], ignore_index=True)
})

sns.boxplot(ax=axes[1], data=teste_data, x="Teste", y="Notas", hue="Grupo")
axes[1].set_title("Pré e Pós-teste por Grupo")
axes[1].set_ylabel("Notas")
axes[1].set_xlabel("Teste")

plt.tight_layout()
plt.show()

# Gráfico 1: Atividades 1, 2 e 3 para os grupos Controle e GPT
fig1, ax1 = plt.subplots(figsize=(8, 6))
sns.boxplot(ax=ax1, data=atividade_data, x="Atividade", y="Notas", hue="Grupo")
ax1.set_title("Atividades 1, 2 e 3 por Grupo")
ax1.set_ylabel("Notas")
ax1.set_xlabel("Atividade")
fig1.tight_layout()
fig1_path = '/boxplot_atividades.png'
fig1.savefig(fig1_path)

# Gráfico 2: Pré-teste e Pós-teste para os grupos Controle e GPT
fig2, ax2 = plt.subplots(figsize=(8, 6))
sns.boxplot(ax=ax2, data=teste_data, x="Teste", y="Notas", hue="Grupo")
ax2.set_title("Pré e Pós-teste por Grupo")
ax2.set_ylabel("Notas")
ax2.set_xlabel("Teste")
fig2.tight_layout()
fig2_path = '/boxplot_teste.png'
fig2.savefig(fig2_path)

fig1_path, fig2_path
```
